## Teste de hipóteses

Queremos descobrir se os futuros clientes, ou seja, a população em geral desse estudo se comportarão dessa mesma maneira com a sobremesa interferindo no valor da gorjeta.

Para solucionar esse problemas usaremos um teste de hipótese


**Hipóteses**

**H<sup>null</sup>:**A distribuição da taxa da gorjeta é a mesma nos dois grupos"
**H<sup>alt</sup>:**A distribuição da taxa da gorjeta não é a mesma nos dois grupos"

```
from scipy.stats import ranksums
#Biblioteca de uso estatístico para realizar teste de hipóteses
```

```
gorjeta.query("sobremesa == 'Sim'").porcentagem
#Pegando do dataframe gorjetas, todas os valores de porcentagem, que tenham sobremesa = sim
```

```
sobremesa = gorjeta.query("sobremesa == 'Sim'").porcentagem
```

```
sem_sobremesa = gorjeta.query("sobremesa == 'Não'").porcentagem
#Pegando do dataframe gorjetas, todas os valores de porcentagem, que tenham sobremesa = não
```

```
r = ranksums(sobremesa, sem_sobremesa)
#O ranksums utiliza o teste de wilcoxon para verificar se existe diferença entre duas situações/amostras
#Utilizados para amostras dependentes
#Ele verifica como hipótese nula se são iguais, como hipótese nula se são diferentes
#Para rejeitar a hipótese nula precisamos de um p-valor menor que 5%
```

```
print('O valor do p-value é {}'.format(r.pvalue))
```

**Conclusão:**

Com um p-valor de 0.52, significa que, apesar de os dois grupos terem dado gorjetas de valores diferentes entre si, a diferença é insignificante do ponto de vista estatístico. Ou seja, não poderemos dizer que a população geral será diferente também.

Verificamos que tem mais pessoas no sábado do que no domingo, podemos ver que a porcentagem entre eles é diferente, será que esses grupos são de fato diferentes?

## Teste de hipóteses

A média do valor da gorjeta no domingo é maior do que a do sábado. Porém, também observamos que a quantidade de pessoas no restaurante é maior no sábado do que no domingo.

A questão é se os valores das contas nos dois dias são diferentes também.

**H<sup>null</null>: A distribuição do valor da conta é igual no sábado e no domingo**

**H<sup>alt</null>: A distribuição do valor da conta não é igual no sábado e no domingo**

```
valor_conta_domingo = gorjeta.query("dia_da_semana == 'Domingo'").valor_da_conta
```

```
valor_conta_sabado = gorjeta.query("dia_da_semana == 'Sábado'").valor_da_conta
```

```
r2 = ranksums(valor_conta_domingo, valor_conta_sabado)
print('O valor do p-value é {}'.format(r2.pvalue))
# A distribução do valor da conta é igual nos dois
```

## Análise estatística

```
gorjeta.groupby(['hora_do_dia']).mean()[['valor_da_conta', 'gorjeta', 'porcentagem']]
```

### Teste de hipóteses

**H<sup>null</null>: A distribuição do valor da conta é igual no jantar e almoço**

**H<sup>alt</null>: A distribuição do valor da conta não é igual no jantar e no almoço**


```
jantar = gorjeta.query("hora_do_dia == 'Jantar'").valor_da_conta #Filtra todos os valores de conta em que hora do dia é jantar
almoco = gorjeta.query("hora_do_dia == 'Almoço'").valor_da_conta #Filtra todos os valores de conta em que hora do dia é almoço
```

```
r3 = ranksums(almoco,jantar)
```

```
print('O valor do p-value é de {}'.format(r3.pvalue))
```

Há indicíos de que a hipótese correta seja: 

**H<sup>alt</null>: A distribuição do valor da conta não é igual no jantar e no almoço**




Ou seja, as distribuições dos valores da conta em horários do dia diferentes não são iguais

Outro teste de hipótese:

**H<sup>null</null>: A distribuição a taxa da gorjeta é igual no jantar e almoço**

**H<sup>alt</null>: A distribuição a taxa da gorjeta não é igual no jantar e no almoço**


```
porcentagem_jantar = gorjeta.query("hora_do_dia == 'Jantar'").porcentagem
```

```
porcentagem_almoco = gorjeta.query("hora_do_dia == 'Almoço'").porcentagem
```

```
r4 = ranksums(porcentagem_almoco, porcentagem_jantar)
```

```
print('O valor do p-value é de {}'.format(r4.pvalue))
```

Há indicíos de que a hipótese correta seja: 

**H<sup>null</null>: A distribuição a taxa da gorjeta é igual no jantar e almoço**



Ou seja, a taxa de gorjeta é igual em ambas partes do dia
